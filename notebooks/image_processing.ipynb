```python
#  Importer les bibliothques
import cv2
import numpy as np
import matplotlib.pyplot as plt
#  Charger l'image (assure-toi que limage est dans le mme dossier que le notebook)
image_path = "tomato_sample.jpg"  # Remplace avec ton image
image = cv2.imread(image_path)
#  Vrifier si limage est bien charge
if image is None:
    print("Erreur : Image non trouve. Vrifie le chemin du fichier.")
else:
    #  Convertir en niveaux de gris
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    #  Appliquer un seuillage (binarisation)
    _, thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY_INV)
    #  Dtection des contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
     #  Dessiner les contours sur l'image originale
    image_contours = image.copy()
    cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)  # Vert
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axs[0].set_title("Image originale")
    axs[0].axis("off")

    axs[1].imshow(gray, cmap="gray")
    axs[1].set_title("Niveaux de gris")
    axs[1].axis("off")

    axs[2].imshow(cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB))
    axs[2].set_title("Contours dtects")
    axs[2].axis("off")

    plt.show()
```
